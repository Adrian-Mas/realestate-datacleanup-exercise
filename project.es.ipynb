#### Ejercicio 15. Traza un gráfico de barras de la mediana de los precios y explica lo que observas (debes usar el subconjunto obtenido del Ejercicio 14) (★★★)

Imprima un gráfico de barras de la mediana de precios y escriba en la celda Markdown un breve análisis sobre el gráfico.

```python
import matplotlib.pyplot as plt

# Calcular la mediana de precios por población
mediana_precios = df_filtrado.groupby('level5')['price'].median().reset_index()
print(mediana_precios)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(mediana_precios['level5'], mediana_precios['price'], color='skyblue')

# Añadir etiquetas a las barras
for index, value in enumerate(mediana_precios['price']):
    plt.text(index, value, str(round(value, 2)), ha='center', va='bottom')

# Configurar el gráfico
plt.title('Mediana de Precios por Población', fontsize=16)
plt.xlabel('Población', fontsize=14)
plt.ylabel('Mediana de Precios', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Me indica que en Getafe el precio medio es mas elevado o hay algunos inmbuebles con valores muy elevados que elevan la mediana

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python

def normalize_price(group):
    return (group - group.mean()) / group.std()


df_filtrado['normalized_price'] = df_filtrado.groupby('level5')['price'].transform(normalize_price)


poblaciones = df_filtrado['level5'].unique()

# Trazar histogramas
plt.figure(figsize=(12, 8))

for poblacion in poblaciones:
    # Filtrar los datos para cada población
    data = df_filtrado[df_filtrado['level5'] == poblacion]['normalized_price']
    plt.hist(data, bins=30, alpha=0.5, label=poblacion)

# Configuración del gráfico
plt.title('Histogramas de Precios Normalizados por Población', fontsize=16)
plt.xlabel('Precio Normalizado', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.legend(title='Poblaciones')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

Alcorcon, Leganés y Fuenlabrada tienen casas de un precio mas bajo y con una mayor oferta, mientras que Getafe polariza mas su oferta, dispone de pisos bastante baratos y pisos mas caros, se alejan de la media.

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
ds['price_per_m2'] = ds['price'] / ds['surface']


poblaciones_interes = ['Arroyomolinos (Madrid)', 'Valdemorillo', 'Galapagar', 'Boadilla del Monte']

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
axs = axs.flatten()  


for i, poblacion in enumerate(poblaciones_interes):

    datos_poblacion = ds[ds['level5'] == poblacion]

    axs[i].scatter(datos_poblacion['surface'], datos_poblacion['price_per_m2'], color='skyblue', alpha=0.6)
    

    axs[i].set_title(f'Diagrama de Dispersión en {poblacion}', fontsize=16)
    axs[i].set_xlabel('Superficie (m²)', fontsize=14)
    axs[i].set_ylabel('Precio por m² (€)', fontsize=14)
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
